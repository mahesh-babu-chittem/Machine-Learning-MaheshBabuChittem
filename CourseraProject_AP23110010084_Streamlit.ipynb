```python
%%writefile app.py
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

st.title("Univariate Analysis of Tweets")

# Load the data
@st.cache_data
def load_data():
    df = pd.read_csv('/content/Tweets.csv')
    return df

df = load_data()

st.header("Dataset Overview")
st.write(df.head())

st.header("Univariate Analysis")

# Select a column for analysis
column = st.selectbox("Select a column to analyze", df.columns)

st.subheader(f"Analysis of {column}")

# Display basic statistics
st.write("Descriptive Statistics:")
st.write(df[column].describe())

# Display distribution plot (if numerical)
if df[column].dtype in ['int64', 'float64']:
    st.subheader("Distribution Plot")
    fig, ax = plt.subplots()
    sns.histplot(df[column], kde=True, ax=ax)
    st.pyplot(fig)

# Display value counts (if categorical)
elif df[column].dtype == 'object':
    st.subheader("Value Counts")
    st.write(df[column].value_counts())
    fig, ax = plt.subplots()
    df[column].value_counts().plot(kind='bar', ax=ax)
    st.pyplot(fig)
```
